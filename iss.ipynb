```python
import requests
import matplotlib.pyplot as plot
import time

data_points = []

for _ in range(100):
    response = requests.get("http://api.open-notify.org/iss-now.json").json()
    data_points.append((float(response['iss_position']['latitude']),
                        float(response['iss_position']['longitude']),
                        response['timestamp']))
    time.sleep(10)

latitudes, longitudes, timestamps = zip(*data_points)

# Scatter plot of the ISS path
plot.scatter(longitudes, latitudes, c = timestamps, cmap ='viridis', marker='.')
plot.xlabel('Longitude')
plot.ylabel('Latitude')
plot.title('ISS Path Tracking')
plot.colorbar(label='Timestamp (expressed in secs)')
plot.show()
```
